```python
from google import genai
from google.colab import userdata
import os
os.environ["GEMINI_API_KEY"] = userdata.get('GEMINI_API_KEY')
geminikey = os.environ["GEMINI_API_KEY"]
client = genai.Client(api_key=geminikey)
```

```python
import pandas as pd

sft_train = pd.read_csv('/content/new_823_data_daihoc.csv')
sft_train
```

```python
sft_train = sft_train.dropna()
sft_train = sft_train.reset_index(drop=True)
sft_train
```

```python
prompt = ""
#for message in input:
#    role = message['role']
#    content = message['content']
#   if role == 'user':
#      prompt += f"[User]\n{content}\n\n"
# elif role == 'assistant':
#    prompt += f"[Original Answer]\n{content}\n\n"
number =1
query = sft_train["Câu hỏi"][number]
ref_context = sft_train["new_ref"][number]
assistant = sft_train["Câu trả lời"][number] + f" Câu trả lời được dựng trên văn bản tham chiếu:'{ref_context}'"
prompt = f"[User]\n{query}\n\n"
prompt += f"[Original Answer]\n{assistant}\n\n"
# Thêm yêu cầu chuyển sang Chain of Thought
prompt += (
    "[Instruction]\nHãy viết lại câu trả lời của assistant ở trên theo dạng Chain of Thought bằng tiếng Việt, "
    "trình bày các bước suy luận rõ ràng trước khi đưa ra kết luận. Trước khi đưa ra kết luận cần có **Câu trả lời**\n"
)

# Gửi yêu cầu
response = client.models.generate_content(
    model="gemma-3-27b-it",  # Replace with the actual model name for Gemini
    contents=prompt
)

# In kết quả
print(response.text)
```

```python
import time
new_ans = []
for i in range(0,sft_train.shape[0]):
  query = sft_train["Câu hỏi"][i]
  ref_context = sft_train["new_ref"][i]
  assistant = sft_train["Câu trả lời"][i] + f" Câu trả lời được dựng trên văn bản tham chiếu:'{ref_context}'"

  prompt = ""
  prompt = f"[User]\n{query}\n\n"
  prompt += f"[Original Answer]\n{assistant}\n\n"
  # Thêm yêu cầu chuyển sang Chain of Thought
  prompt += (
      "[Instruction]\nHãy viết lại câu trả lời của assistant ở trên theo dạng Chain of Thought bằng tiếng Việt, "
      "trình bày các bước suy luận rõ ràng trước khi đưa ra kết luận. Trước khi đưa ra kết luận cần có **Câu trả lời**\n"
  )
  response = client.models.generate_content(
    model="gemma-3-27b-it",  # Replace with the actual model name for Gemini
    contents=prompt
  )

  # In kết quả
  print(i)
  new_ans.append(response.text)
  #time.sleep(2)  # Giới hạn 15 request/phút
```
